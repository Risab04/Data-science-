```python
import random

population =list(range(1,1000))
sample_size = 10
random_sample = random.sample(population,sample_size)
print("random sample", random_sample )
```

```python
import numpy as np
import pandas as pd

data = pd.DataFrame({
    'category' : ['a','b','c','d']*25 ,
    'value': np.random.randn(100)
})

stratified_sample = data.groupby('category',group_keys = False).apply(lambda x:x.sample(5))

print(stratified_sample)


```

```python
clusters = [list(range(1,26)),list(range(26,51)),list(range(51,76)),list(range(101,201))]
cluster_sample = random.choice(clusters)
print(cluster_sample)
```

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'values': np.random.randn(101)
})

summary = data.describe()
print(summary)
```

```python
import matplotlib.pyplot as plt
data['values'].hist(bins=5000,edgecolor='red')
plt.title('histogram')
plt.xlabel('value')
plt.ylabel('freq')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(iris.data)

plt.scatter(reduced_data[:,0],reduced_data[:,1],c=iris.target ,cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.title('pca of iris data sets ')
plt.colorbar(label='species')
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/iris-data.csv')
parallel_coordinates(data, 'class',color=['b','g','r'])
plt.title('parallel coordinates plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
